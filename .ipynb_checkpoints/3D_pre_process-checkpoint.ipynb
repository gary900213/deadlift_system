```python
##keypoints3d轉angle.txt
import os
import json
import math
import numpy as np

def calculate_angle(a, b, c):
    ba = [a[i] - b[i] for i in range(3)]
    bc = [c[i] - b[i] for i in range(3)]
    dot_product = sum(ba[i] * bc[i] for i in range(3))
    magnitude_ba = math.sqrt(sum(ba[i] ** 2 for i in range(3)))
    magnitude_bc = math.sqrt(sum(bc[i] ** 2 for i in range(3)))
    if magnitude_ba == 0 or magnitude_bc == 0:
        return 0.0
    cos_angle = dot_product / (magnitude_ba * magnitude_bc)
    cos_angle = max(min(cos_angle, 1.0), -1.0)
    angle_rad = math.acos(cos_angle)
    return math.degrees(angle_rad)

def calculate_distance(a, b):
    return math.sqrt(sum((a[i] - b[i]) ** 2 for i in range(3)))

def vector_angle(a, b):
    dot = sum(a[i] * b[i] for i in range(3))
    mag_a = math.sqrt(sum(a[i] ** 2 for i in range(3)))
    mag_b = math.sqrt(sum(b[i] ** 2 for i in range(3)))
    if mag_a == 0 or mag_b == 0:
        return 0.0
    cos_val = max(min(dot / (mag_a * mag_b), 1.0), -1.0)
    return math.degrees(math.acos(cos_val))

# 索引
NECK = 1
L_SHOULDER, L_ELBOW = 5, 6
R_SHOULDER, R_ELBOW = 2, 3
L_WRIST, R_WRIST = 7, 4
L_HIP, R_HIP = 12, 9
L_KNEE, R_KNEE = 13, 10
L_ANKLE, R_ANKLE = 14, 11

# 主資料夾
for i in range(5):
    base_folder = f"D:\\labdata\\MOCAP\\1234\\Category_{i+1}"
    
    # 遍歷所有子資料夾
    for subfolder in os.listdir(base_folder):
        subfolder_path = os.path.join(base_folder, subfolder, "keypoints3d")
        if not os.path.isdir(subfolder_path):
            continue
    
        angle_file_path = os.path.join(base_folder, subfolder, "angle.txt")
        with open(angle_file_path, "w") as f_out:
            for filename in sorted(os.listdir(subfolder_path)):
                if filename.endswith(".json"):
                    frame_number = int(filename.split(".")[0])
                    file_path = os.path.join(subfolder_path, filename)
    
                    with open(file_path, 'r') as f_json:
                        data = json.load(f_json)
                        if not data or 'keypoints3d' not in data[0]:
                            continue
                        kp = data[0]['keypoints3d']
                        if len(kp) < 25:
                            continue
    
                        neck = kp[NECK][:3]
                        l_shoulder, l_elbow, l_wrist = kp[L_SHOULDER][:3], kp[L_ELBOW][:3], kp[L_WRIST][:3]
                        r_shoulder, r_elbow, r_wrist = kp[R_SHOULDER][:3], kp[R_ELBOW][:3], kp[R_WRIST][:3]
                        l_hip, r_hip = kp[L_HIP][:3], kp[R_HIP][:3]
                        l_knee, r_knee = kp[L_KNEE][:3], kp[R_KNEE][:3]
                        l_ankle, r_ankle = kp[L_ANKLE][:3], kp[R_ANKLE][:3]
    
                        left_knee_angle = calculate_angle(l_hip, l_knee, l_ankle)
                        right_knee_angle = calculate_angle(r_hip, r_knee, r_ankle)
                        left_hip_angle = calculate_angle(l_shoulder, l_hip, l_knee)
                        right_hip_angle = calculate_angle(r_shoulder, r_hip, r_knee)
    
                        mid_hip = [(l_hip[i] + r_hip[i]) / 2 for i in range(3)]
                        body_length = calculate_distance(neck, mid_hip)
    
                        body_vec = [mid_hip[i] - neck[i] for i in range(3)]
                        l_arm_vec = [l_elbow[i] - l_shoulder[i] for i in range(3)]
                        r_arm_vec = [r_elbow[i] - r_shoulder[i] for i in range(3)]
    
                        left_arm_body_angle = vector_angle(body_vec, l_arm_vec)
                        right_arm_body_angle = vector_angle(body_vec, r_arm_vec)
    
                        f_out.write(f"{frame_number} "
                                    f"{left_knee_angle:.4f} {right_knee_angle:.4f} "
                                    f"{left_hip_angle:.4f} {right_hip_angle:.4f} "
                                    f"{body_length:.4f} "
                                    f"{left_arm_body_angle:.4f} {right_arm_body_angle:.4f}\n")

        print(f"✅ 完成處理：{subfolder}")
```

```python
##畫angle.txt(去極值)
import matplotlib.pyplot as plt
import numpy as np

# 檔案路徑
angle_file = "D:\\labdata\\MOCAP\\1234\\Category_1\\recording_20241212_103521\\angle.txt"

# 初始化資料
frames = []
left_knee_angles = []
right_knee_angles = []
left_hip_angles = []
right_hip_angles = []
body_lengths = []
left_arm_angles = []
right_arm_angles = []

# 讀取資料
with open(angle_file, "r") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 8:
            frame = int(parts[0])
            angles = list(map(float, parts[1:]))

            frames.append(frame)
            left_knee_angles.append(angles[0])
            right_knee_angles.append(angles[1])
            left_hip_angles.append(angles[2])
            right_hip_angles.append(angles[3])
            body_lengths.append(angles[4])
            left_arm_angles.append(angles[5])
            right_arm_angles.append(angles[6])

# ---------- 去除離群值並進行內插補值 ----------

def remove_outliers_and_interpolate(data, min_valid_value=1e-3):
    data = np.array(data, dtype=np.float64)

    # 先把明顯錯誤（如 0 或負數）的值先標為 NaN
    data[data < min_valid_value] = np.nan

    # 接著再計算 mean/std
    mean = np.nanmean(data)
    std = np.nanstd(data)
    lower_bound = mean - 1.5 * std
    upper_bound = mean + 1.5* std

    # 找出離群值（大於 upper_bound 或小於 lower_bound）
    outliers = (data < lower_bound) | (data > upper_bound)
    data[outliers] = np.nan

    # 線性內插補值
    indices = np.arange(len(data))
    not_nan = ~np.isnan(data)
    data_interp = np.copy(data)
    data_interp[np.isnan(data)] = np.interp(indices[np.isnan(data)], indices[not_nan], data[not_nan])

    return data_interp, lower_bound, upper_bound


# ----------- 準備繪圖資料 -----------

left_knee_angles, lk_lb, lk_ub = remove_outliers_and_interpolate(left_knee_angles)
right_knee_angles, rk_lb, rk_ub = remove_outliers_and_interpolate(right_knee_angles)
left_hip_angles, lh_lb, lh_ub = remove_outliers_and_interpolate(left_hip_angles)
right_hip_angles, rh_lb, rh_ub = remove_outliers_and_interpolate(right_hip_angles)
body_lengths, bl_lb, bl_ub = remove_outliers_and_interpolate(body_lengths)
left_arm_angles, la_lb, la_ub = remove_outliers_and_interpolate(left_arm_angles)
right_arm_angles, ra_lb, ra_ub = remove_outliers_and_interpolate(right_arm_angles)

# ----------- 開始繪圖 -----------

# 膝蓋角度
plt.figure(figsize=(10, 4))
plt.plot(frames, left_knee_angles, label="Left Knee Angle")
plt.plot(frames, right_knee_angles, label="Right Knee Angle")
plt.axhline(lk_lb, color='red', linestyle='--', label="Left Knee Lower Bound")
plt.axhline(lk_ub, color='red', linestyle='--', label="Left Knee Upper Bound")
plt.axhline(rk_lb, color='orange', linestyle='--', label="Right Knee Lower Bound")
plt.axhline(rk_ub, color='orange', linestyle='--', label="Right Knee Upper Bound")
plt.title("Knee Angles Over Time")
plt.xlabel("Frame")
plt.ylabel("Angle (°)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 髖關節角度
plt.figure(figsize=(10, 4))
plt.plot(frames, left_hip_angles, label="Left Hip Angle")
plt.plot(frames, right_hip_angles, label="Right Hip Angle")
plt.axhline(lh_lb, color='red', linestyle='--', label="Left Hip Lower Bound")
plt.axhline(lh_ub, color='red', linestyle='--', label="Left Hip Upper Bound")
plt.axhline(rh_lb, color='orange', linestyle='--', label="Right Hip Lower Bound")
plt.axhline(rh_ub, color='orange', linestyle='--', label="Right Hip Upper Bound")
plt.title("Hip Angles Over Time")
plt.xlabel("Frame")
plt.ylabel("Angle (°)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 身體長度
plt.figure(figsize=(10, 4))
plt.plot(frames, body_lengths, label="Body Length", color='purple')
plt.axhline(bl_lb, color='red', linestyle='--', label="Lower Bound")
plt.axhline(bl_ub, color='red', linestyle='--', label="Upper Bound")
plt.title("Body Length Over Time")
plt.xlabel("Frame")
plt.ylabel("Length")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 手臂與身體夾角
plt.figure(figsize=(10, 4))
plt.plot(frames, left_arm_angles, label="Left Arm-Body Angle", color='green')
plt.plot(frames, right_arm_angles, label="Right Arm-Body Angle", color='orange')
plt.axhline(la_lb, color='green', linestyle='--', label="Left Arm Lower Bound")
plt.axhline(la_ub, color='green', linestyle='--', label="Left Arm Upper Bound")
plt.axhline(ra_lb, color='orange', linestyle='--', label="Right Arm Lower Bound")
plt.axhline(ra_ub, color='orange', linestyle='--', label="Right Arm Upper Bound")
plt.title("Arm-Body Angles Over Time")
plt.xlabel("Frame")
plt.ylabel("Angle (°)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Hampel濾波器
import os
import numpy as np
import matplotlib.pyplot as plt

# 資料根目錄
base_dir = "D:\\labdata\\MOCAP\\1234"

# Hampel濾波器 + 線性內插
def hampel_filter_interpolate(data, window_size=10, n_sigma=3, min_valid_value=1e-3):
    data = np.array(data, dtype=np.float64)
    data[data < min_valid_value] = np.nan  # 小於 min_valid_value 視為異常，設為 NaN

    n = len(data)
    filtered = np.copy(data)

    for i in range(n):
        window_start = max(i - window_size, 0)
        window_end = min(i + window_size + 1, n)
        window = data[window_start:window_end]

        median = np.nanmedian(window)
        mad = np.nanmedian(np.abs(window - median))  # median absolute deviation

        if mad == 0:
            threshold = 0
        else:
            threshold = n_sigma * 1.4826 * mad

        if not np.isnan(data[i]) and np.abs(data[i] - median) > threshold:
            filtered[i] = np.nan  # 離群值設為 NaN

    # 線性內插 NaN
    indices = np.arange(len(filtered))
    not_nan = ~np.isnan(filtered)
    filtered[np.isnan(filtered)] = np.interp(indices[np.isnan(filtered)], indices[not_nan], filtered[not_nan])

    return filtered

# 取得 bar 標記區間
def get_bar_ranges(bar_dir):
    bar_ranges = []

    if not os.path.isdir(bar_dir):
        return bar_ranges

    for file in os.listdir(bar_dir):
        if file.endswith(".txt"):
            path = os.path.join(bar_dir, file)
            try:
                with open(path, "r") as f:
                    frames = []
                    for line in f:
                        line = line.strip()
                        if not line:
                            continue
                        parts = line.split(",")
                        if parts:
                            frame = int(parts[0].strip())
                            frames.append(frame)
                    if frames:
                        start = min(frames)
                        end = max(frames)
                        if start < end:
                            bar_ranges.append((start, end))
            except Exception as e:
                print(f"讀取 bar file 時錯誤: {path}, {e}")

    return bar_ranges

# 繪圖
def plot_comparison(frames, raw, cleaned, save_path, title, bar_ranges):
    plt.figure(figsize=(10, 4))
    plt.plot(frames, raw, label="Raw", color='gray', alpha=0.6)
    plt.plot(frames, cleaned, label="Cleaned", color='blue')

    # 顏色輪替
    colors = ['yellow', 'orange', 'lightgreen', 'lightblue', 'pink', 'lightcoral']
    for idx, (start, end) in enumerate(bar_ranges):
        color = colors[idx % len(colors)]
        plt.axvspan(start, end, color=color, alpha=0.3, label=f'Bar {idx+1}' if idx < len(colors) else None)

    plt.title(title)
    plt.xlabel("Frame")
    plt.ylabel("Angle / Length")

    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())

    plt.grid(True)
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()

# 主程序
for category in range(1, 6):
    category_path = os.path.join(base_dir, f"Category_{category}")
    if not os.path.isdir(category_path):
        continue

    for recording in os.listdir(category_path):
        recording_path = os.path.join(category_path, recording)
        angle_file = os.path.join(recording_path, "angle_int.txt")
        if not os.path.isfile(angle_file):
            continue

        print(f"處理中: {angle_file}")

        # 讀取資料
        frames = []
        angle_lists = [[] for _ in range(7)]
        with open(angle_file, "r") as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) == 8:
                    frames.append(int(parts[0]))
                    for i in range(7):
                        angle_lists[i].append(float(parts[i+1]))

        # 取得 bar 標記
        bar_folder = os.path.join(recording_path, "bar")
        bar_ranges = get_bar_ranges(bar_folder)

        # 處理每個角度
        processed_angles = []
        for i, angle in enumerate(angle_lists):
            cleaned = hampel_filter_interpolate(angle)
            processed_angles.append(cleaned)

            # 繪圖
            plot_path = os.path.join(recording_path, f"angle_compare_hampel_{i+1}.png")
            plot_comparison(frames, angle, cleaned, plot_path, f"Angle {i+1} Comparison", bar_ranges)

        # 存清理後的檔案
        output_path = os.path.join(recording_path, "angle_int_hampel.txt")
        with open(output_path, "w") as out_f:
            for i, frame in enumerate(frames):
                angle_values = [f"{processed_angles[j][i]:.4f}" for j in range(7)]
                out_f.write(f"{frame} " + " ".join(angle_values) + "\n")

        print(f"完成: {output_path}")
```

```python
##處理整體極值與刪0
import os
import numpy as np
import matplotlib.pyplot as plt

# 資料根目錄
base_dir = "D:\\labdata\\MOCAP\\1234"

# 去除離群值並進行內插補值（排除0等不合理值）
def remove_outliers_and_interpolate(data, min_valid_value=1e-3):
    data = np.array(data, dtype=np.float64)
    data[data < min_valid_value] = np.nan
    mean = np.nanmean(data)
    std = np.nanstd(data)
    lower_bound = mean - 2.2 * std
    upper_bound = mean + 2.2 * std
    outliers = (data < lower_bound) | (data > upper_bound)
    data[outliers] = np.nan
    indices = np.arange(len(data))
    not_nan = ~np.isnan(data)
    data_interp = np.copy(data)
    data_interp[np.isnan(data)] = np.interp(indices[np.isnan(data)], indices[not_nan], data[not_nan])
    return data_interp


# 取得 bar 標記區間（依據每個 bar txt 裡的幀數範圍）
def get_bar_ranges(bar_dir):
    bar_ranges = []

    if not os.path.isdir(bar_dir):
        return bar_ranges

    for file in os.listdir(bar_dir):
        if file.endswith(".txt"):
            path = os.path.join(bar_dir, file)
            try:
                with open(path, "r") as f:
                    frames = []
                    for line in f:
                        line = line.strip()
                        if not line:
                            continue
                        parts = line.split(",")
                        if parts:
                            frame = int(parts[0].strip())  # 抓第一欄，並去除空白
                            frames.append(frame)
                    if frames:
                        start = min(frames)
                        end = max(frames)
                        if start < end:
                            bar_ranges.append((start, end))
            except Exception as e:
                print(f"讀取 bar file 時錯誤: {path}, {e}")

    return bar_ranges




# 繪製原始 vs 清理後圖表，並標註 bar 區塊
def plot_comparison(frames, raw, cleaned, save_path, title, bar_ranges):
    plt.figure(figsize=(10, 4))
    plt.plot(frames, raw, label="Raw", color='gray', alpha=0.6)
    plt.plot(frames, cleaned, label="Cleaned", color='blue')
    print(bar_ranges) 
    # 顯示 bar 標記區間
    for start, end in bar_ranges:
        plt.axvspan(start, end, color='yellow', alpha=0.3)

    plt.title(title)
    plt.xlabel("Frame")
    plt.ylabel("Angle / Length")
    plt.legend()
    plt.grid(True)

    
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()
def plot_comparison(frames, raw, cleaned, save_path, title, bar_ranges):
    plt.figure(figsize=(10, 4))
    plt.plot(frames, raw, label="Raw", color='gray', alpha=0.6)
    plt.plot(frames, cleaned, label="Cleaned", color='blue')

    # 顏色列表，輪流用
    colors = ['yellow', 'orange', 'lightgreen', 'lightblue', 'pink', 'lightcoral']
    
    for idx, (start, end) in enumerate(bar_ranges):
        color = colors[idx % len(colors)]
        plt.axvspan(start, end, color=color, alpha=0.3, label=f'Bar Segment {idx+1}' if idx < len(colors) else None)

    plt.title(title)
    plt.xlabel("Frame")
    plt.ylabel("Angle / Length")
    
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())  # 避免重複 label
    
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()

# 遍歷資料夾
for category in range(1, 6):
    category_path = os.path.join(base_dir, f"Category_{category}")
    if not os.path.isdir(category_path):
        continue

    for recording in os.listdir(category_path):
        recording_path = os.path.join(category_path, recording)
        angle_file = os.path.join(recording_path, "angle.txt")
        if not os.path.isfile(angle_file):
            continue

        print(f"處理中: {angle_file}")
        
        # 讀取原始資料
        frames = []
        angle_lists = [[] for _ in range(7)]
        with open(angle_file, "r") as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) == 8:
                    frames.append(int(parts[0]))
                    for i in range(7):
                        angle_lists[i].append(float(parts[i+1]))

        # bar 區塊
        bar_folder = os.path.join(recording_path, "bar")
        bar_ranges = get_bar_ranges(bar_folder)

        # 處理每個角度，並保存圖表
        processed_angles = []
        for i, angle in enumerate(angle_lists):
            cleaned = remove_outliers_and_interpolate(angle)
            processed_angles.append(cleaned)

            # 繪製原始與清理後對比圖
            plot_path = os.path.join(recording_path, f"angle_compare_{i+1}.png")
            plot_comparison(frames, angle, cleaned, plot_path, f"Angle {i+1} Comparison", bar_ranges)

        # 儲存清理後資料
        output_path = os.path.join(recording_path, "angle_int.txt")
        with open(output_path, "w") as out_f:
            for i, frame in enumerate(frames):
                angle_values = [f"{processed_angles[j][i]:.4f}" for j in range(7)]
                out_f.write(f"{frame} " + " ".join(angle_values) + "\n")

        print(f"完成: {output_path}")
```

```python
##把angle.txt按2d bar的幀數來切割影像
import os
import pandas as pd

root_dir = "D:\\labdata\\MOCAP\\1234"
output_folder_name = "angle_int"

for category in os.listdir(root_dir):
    category_path = os.path.join(root_dir, category)
    if not os.path.isdir(category_path):
        continue

    for recording in os.listdir(category_path):
        recording_path = os.path.join(category_path, recording)
        bar_path = os.path.join(recording_path, "bar")
        angle_path = os.path.join(recording_path, "angle_int_hampel.txt")
        output_path = os.path.join(recording_path, output_folder_name)

        if not os.path.exists(bar_path) or not os.path.isfile(angle_path):
            continue

        # 讀 angle.txt（用空格分隔）
        angle_df = pd.read_csv(angle_path, header=None, delim_whitespace=True)

        # 創建輸出資料夾
        os.makedirs(output_path, exist_ok=True)

        # 遍歷每個 bar/*.txt
        for file_name in sorted(os.listdir(bar_path)):
            if not file_name.endswith(".txt"):
                continue
            bar_file = os.path.join(bar_path, file_name)

            # 讀 bar 檔：取第一列與最後一列的第一個值（用逗號分隔）
            with open(bar_file, "r") as f:
                lines = f.readlines()
                if not lines:
                    continue
                start_frame = int(lines[0].strip().split(",")[0])
                end_frame = int(lines[-1].strip().split(",")[0])

            # 從 angle_df 中擷取對應幀數區間
            sub_angle_df = angle_df[(angle_df[0] >= start_frame) & (angle_df[0] <= end_frame)]

            # 存成新檔案（用逗號分隔）
            angle_output_name = f"angle_{os.path.splitext(file_name)[0].split('_')[-1]}.txt"
            sub_angle_df.to_csv(os.path.join(output_path, angle_output_name), index=False, header=False, sep=',')


```

```python

import cv2
import os
from natsort import natsorted  # 用於自然排序檔名（如000001.jpg, 000002.jpg）

# 設定圖片所在資料夾路徑
image_folder = 'D:\\labdata\\MOCAP\\EasyMocap-master\\0_3Dpose_es\\subject14\\recording_20241205_134801\\repro'  # 例如：'./images'
output_video_path ='D:\\labdata\\MOCAP\\EasyMocap-master\\0_3Dpose_es\\subject14\\recording_20241205_134801\\repro\\output_video.mp4'
fps = 30  # 每秒幾張

# 取得所有圖片檔名並排序
images = [img for img in os.listdir(image_folder) if img.endswith(".jpg")]
images = natsorted(images)

# 讀取第一張圖片取得尺寸
first_image_path = os.path.join(image_folder, images[0])
frame = cv2.imread(first_image_path)
height, width, layers = frame.shape

# 建立影片寫入器
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 使用 mp4 編碼
video = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# 將每張圖片寫入影片
for image in images:
    img_path = os.path.join(image_folder, image)
    frame = cv2.imread(img_path)
    if frame is not None:
        video.write(frame)

video.release()
print("✅ 成功輸出影片：", output_video_path)

```
