```python
import math
import os

def calculate_distance(x1, y1, x2, y2):
    """
    計算兩點之間的歐氏距離
    """
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def calculate_angle(x1, y1, x2, y2, x3, y3):
    """
    計算由三點構成的夾角，x2, y2 是頂點
    """
    c = calculate_distance(x2, y2, x3, y3)
    a = calculate_distance(x1, y1, x3, y3)
    b = calculate_distance(x1, y1, x2, y2)

    if b == 0 or c == 0:
        return 0.0

    try:
        cos_theta = (b**2 + c**2 - a**2) / (2 * b * c)
        cos_theta = max(-1.0, min(1.0, cos_theta))
        angle = math.acos(cos_theta)
        return math.degrees(angle)
    except ValueError:
        return 0.0

def process_file(input_file, output_file):
    joint_data = {}

    try:
        # 讀取檔案並整理資料
        with open(input_file, 'r') as f:
            for line in f:
                frame, index, x, y = map(float, line.strip().split(','))
                frame = int(frame)
                index = int(index)
                if frame not in joint_data:
                    joint_data[frame] = {}
                joint_data[frame][index] = (x, y)

        # 計算角度和長度，並保存結果
        with open(output_file, 'w') as f_out:
            f_out.write("frame,hip_angle,knee_angle,body_length,knee_hip_ratio\n")
            for frame, joints in sorted(joint_data.items()):
                if all(idx in joints for idx in [6, 12, 14, 16]):
                    x6, y6 = joints[6]
                    x12, y12 = joints[12]
                    x14, y14 = joints[14]
                    x16, y16 = joints[16]

                    hip_angle = calculate_angle(x6, y6, x12, y12, x14, y14)
                    knee_angle = calculate_angle(x12, y12, x14, y14, x16, y16)
                    body_length = calculate_distance(x6, y6, x12, y12)
                    knee_hip_ratio = knee_angle / hip_angle if hip_angle != 0 else 0

                    f_out.write(f"{frame},{hip_angle:.2f},{knee_angle:.2f},{body_length:.2f},{knee_hip_ratio:.2f}\n")
    except Exception as e:
        print(f"處理檔案時發生錯誤：{input_file}，錯誤訊息：{e}")

def main():
    input_folder = "D:\\labdata\\MOCAP\\recordings_copy\\受試者52\\recording_20241223_180418\\skeleton"
    output_folder = os.path.join(input_folder, "processed")
    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        if filename.startswith("skeleton_") and filename.endswith(".txt"):
            input_file = os.path.join(input_folder, filename)
            output_file = os.path.join(output_folder, f"processed_{filename}")
            process_file(input_file, output_file)
            print(f"已處理檔案：{input_file}，結果保存至：{output_file}")
            
#     base_path = "D:\\labdata\\MOCAP\\recordings_copy\\受試者10\\"
#     # 取得 base_path 下的所有子資料夾
#     for sub_folder in os.listdir(base_path):
#         path = os.path.join(base_path, sub_folder)
        
#         final_path = path + "\\skeleton"
#         output_folder = os.path.join(final_path, "processed")
#         # 確保 path 是資料夾
#         if not os.path.isdir(final_path):
#             continue
        
#         for filename in os.listdir(final_path):
#             if filename.startswith("skeleton_") and filename.endswith(".txt"):
#                 input_file = os.path.join(final_path, filename)
#                 output_file = os.path.join(output_folder, f"processed_{filename}")
#                 process_file(final_path, output_file)
#                 print(f"已處理檔案：{input_file}，結果保存至：{output_file}")
            
            
            

if __name__ == "__main__":
    main()
```

```python
import math
import os

def calculate_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def calculate_angle(x1, y1, x2, y2, x3, y3):
    c = calculate_distance(x2, y2, x3, y3)
    a = calculate_distance(x1, y1, x3, y3)
    b = calculate_distance(x1, y1, x2, y2)
    if b == 0 or c == 0:
        return 0.0
    try:
        cos_theta = (b**2 + c**2 - a**2) / (2 * b * c)
        cos_theta = max(-1.0, min(1.0, cos_theta))
        return math.degrees(math.acos(cos_theta))
    except ValueError:
        return 0.0

def process_file(input_file, output_file):
    joint_data = {}
    try:
        with open(input_file, 'r') as f:
            for line in f:
                frame, index, x, y = map(float, line.strip().split(','))
                frame = int(frame)
                index = int(index)
                if frame not in joint_data:
                    joint_data[frame] = {}
                joint_data[frame][index] = (x, y)

        with open(output_file, 'w') as f_out:
            for frame, joints in sorted(joint_data.items()): 
                if all(idx in joints for idx in [6, 10, 12, 14, 16]):
                    x6, y6 = joints[6]
                    x10, y10 = joints[10]
                    x12, y12 = joints[12]
                    x14, y14 = joints[14]
                    x16, y16 = joints[16]

                    hip_angle = calculate_angle(x6, y6, x12, y12, x14, y14)
                    knee_angle = calculate_angle(x12, y12, x14, y14, x16, y16)
                    body_length = calculate_distance(x6, y6, x12, y12)
                    arm_angle = calculate_angle(x10, y10, x6, y6, x12, y12)

                    f_out.write(f"{frame},{hip_angle:.2f},{knee_angle:.2f},{body_length:.2f},{arm_angle:.2f}\n")
    except Exception as e:
        print(f"處理檔案時發生錯誤：{input_file}，錯誤訊息：{e}")

def process_all_folders(base_path, target_folder_name="skeleton_vision1"):
    for root, dirs, files in os.walk(base_path):
        if os.path.basename(root) == target_folder_name:
            for filename in files:
                if filename.startswith("skeleton_") and filename.endswith(".txt"):
                    input_file = os.path.join(root, filename)
                    
                    parent_folder = os.path.dirname(root)
                    output_folder = os.path.join(parent_folder, "angle_vision1")
                    os.makedirs(output_folder, exist_ok=True)

                    output_file = os.path.join(output_folder, f"angle_{filename}")
                    process_file(input_file, output_file)
                    print(f"已處理檔案：{input_file}，結果保存至：{output_file}")

def main():
    base_path = "D:\\labdata\\MOCAP\\recordings_copy\\"
    process_all_folders(base_path, target_folder_name="skeleton_vision1")

if __name__ == "__main__":
    main()
```

```python
import os
import re

def extract_number(filename):
    match = re.search(r'(\d+)', filename)
    return match.group(1) if match else None

def process_file(skeleton_file, bar_file, output_file):
    try:
        with open(skeleton_file, 'r') as skel_f, open(bar_file, 'r') as bar_f:
            skel_lines = skel_f.readlines()
            bar_lines = bar_f.readlines()

            # 只保留 index == 5 的 skeleton 行
            shoulder_lines = []
            for line in skel_lines:
                parts = line.strip().split(',')
                if len(parts) >= 4:
                    try:
                        index = int(float(parts[1]))
                        if index == 5:
                            shoulder_lines.append(parts)
                    except ValueError:
                        continue

            pair_count = min(len(shoulder_lines), len(bar_lines))

            with open(output_file, 'w') as out_f:
                for i in range(pair_count):
                    skel_parts = shoulder_lines[i]
                    bar_parts = bar_lines[i].strip().split(',')

                    if len(bar_parts) < 2:
                        continue
                    try:
                        frame = int(float(skel_parts[0]))
                        shoulder_x = float(skel_parts[2])
                        bar_x = float(bar_parts[1])
                        diff = shoulder_x - bar_x
                        out_f.write(f"{frame},{diff:.4f}\n")
                    except ValueError:
                        continue

    except Exception as e:
        print(f"處理檔案錯誤：{skeleton_file}，錯誤：{e}")

def process_all(base_path, target_folder_name="skeleton_vision1"):
    for root, dirs, files in os.walk(base_path):
        if os.path.basename(root) == target_folder_name:
            bar_folder = os.path.join(os.path.dirname(root), "bar")
            output_folder = os.path.join(os.path.dirname(root), "angle_vision1")
            os.makedirs(output_folder, exist_ok=True)

            for filename in files:
                if filename.endswith(".txt") and filename.startswith("skeleton_"):
                    number = extract_number(filename)
                    if number:
                        skel_file = os.path.join(root, filename)
                        bar_file = os.path.join(bar_folder, f"bar_{number}.txt")
                        output_file = os.path.join(output_folder, f"angle_{filename}")

                        if os.path.exists(bar_file):
                            process_file(skel_file, bar_file, output_file)
                            print(f"✅ 已處理：{filename} 對應 bar_{number}.txt")
                        else:
                            print(f"❌ 缺少 bar 對應檔案：bar_{number}.txt")

def main():
    base_path = "D:\\labdata\\MOCAP\\recordings_copy\\"
    process_all(base_path, target_folder_name="skeleton_vision1")

if __name__ == "__main__":
    main()
```
